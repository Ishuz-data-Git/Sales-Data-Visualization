# Project : "Sales Data Visualization Dashboard"

```python
import pandas as pd
```

```python
data = {
    "Month": ['Jan','Feb','Mar','Apr','May','Jun'],
    "Sales": [25000, 27000, 30000, 28000, 35000, 37000],
    "Profit": [4000, 4500, 5000, 4200, 5800, 6000],
    "Expenses": [21000, 22500, 25000, 23800, 29200, 31000]
}
df = pd.DataFrame(data)
print(df)
```

### Line Chart - trend samajhane ke liye

```python
import matplotlib.pyplot as plt

# Task - Show sales trend over month 
plt.plot( df['Month'],df['Sales'], marker='o', label='Sales')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales in Rs.')
plt.legend()
plt.show()
```

### Bar Chart - comparison ke liye

```python
# Task - Compare Profit & Expense
import numpy as np

x = np.arange(len(df['Month'])) # ye array banaega [0,1,2,3,4,5] months ke liye
plt.bar(x - 0.2, df['Profit'], width=0.4, label='Profit', color='black') # x - 0.2 means bar ko left side shift krega, 0.4 bar ki width
plt.bar(x + 0.2, df['Expenses'], width=0.4, label='Expenses', color='yellow') # x + 0.2 ye bar right me banega (compare karne ke liye)
plt.xticks(x, df['Month']) # X-axis pe number (0,1,2..) ke badle month names show krega

plt.title('Profit vs Expenses')
plt.xlabel('Month')
plt.ylabel('Amount (Rs.)')
plt.legend()
plt.show()
```

### Pie Chart - show percentage contribution of sales

```python
plt.pie(df['Sales'], labels=df['Month'],
        autopct='%1.1f%%', startangle=140)
plt.title('Monthly Sales Contribution')
plt.show()
```

### Scatter Plot - Visualize relation between Sales and Profit

```python
plt.scatter(df['Sales'], df['Profit'], color='green')
plt.title('Sales vs Profit')
plt.xlabel('Sales (Rs.)')
plt.ylabel('Profit (Rs.)')
plt.show()
```

### Subplots (Dashboard view) - Show all in one figure

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8)) # 2 row X 2 columns

# Top-left (Line Chart)
axs[0,0].plot(df['Month'], df['Sales'], marker='o', color='blue')
axs[0,0].set_title('Sales Trend')
# Top-right (Bar Chart)
axs[0,1].bar(df['Month'], df['Profit'], color='orange')
axs[0,1].set_title('Profit Per Month')
# Bottom-left (Pie Chart)
axs[1,0].pie(df['Sales'], labels=df['Month'],autopct='%1.1f%%')
axs[1,0].set_title('Sales Contribution')
# Bottom-right (Scatter Plot)
axs[1,1].scatter(df['Sales'], df['Expenses'], color='red')
axs[1,1].set_title('Sales vs Expenses')

plt.tight_layout() 
plt.show()
```
